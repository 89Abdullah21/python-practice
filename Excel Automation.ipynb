```python
import openpyxl as xl # importing the excel module to read and write excel files
wb = xl.load_workbook('Book1.xlsx', data_only=True) # objected created to load the workbook or excel file
sheet = wb['Sheet1'] # Passing the particular sheet name to the object created
cell = sheet['a2'] # Accessing or pointing to the particular cell in the sheet
cell2 = sheet.cell(2,3) # First is row and second is column number
print(cell.value) # To get the value of the cell
print(cell2.value)
```

```python
print(sheet.max_row) # To get the maximum number of rows in the sheet
print(sheet.max_column , '\n\n') # To get the maximum number of columns in the sheet

# The below code is to copy the values from column D to column G in the same sheet

for rows in range(2, sheet.max_row + 1): # Looping through the rows of the sheet
    cell = sheet.cell(rows, 5)
    # print(cell.value)
    correct_values = sheet.cell(rows, 4).value
    correct_values_cell = sheet.cell(rows, 7)
    correct_values_cell.value = correct_values
    print(correct_values_cell.value)

wb.save('Book2.xlsx')
```

```python
# The below code is to create a bar chart for the values in column G

from openpyxl.chart import BarChart, Reference

selected_values = Reference(sheet, min_row=2, max_row=sheet.max_row, min_col=sheet.max_column, max_col=sheet.max_column)
chart = BarChart()
chart.add_data(selected_values)
sheet.add_chart(chart, 'd13')

wb.save('Book3.xlsx')
```

```python
# Making the code clean and resuable by having it in a function

import openpyxl as xl
from openpyxl.chart import BarChart, Reference

def handel_excel(filename):
    wb = xl.load_workbook(filename, data_only=True)
    sheet = wb['Sheet1']

    for rows in range(2, sheet.max_row + 1): # Looping through the rows of the sheet
        cell = sheet.cell(rows, 5)
        # print(cell.value)
        correct_values = sheet.cell(rows, 4).value
        correct_values_cell = sheet.cell(rows, 7)
        correct_values_cell.value = correct_values
        print(correct_values_cell.value)

    from openpyxl.chart import BarChart, Reference

    selected_values = Reference(sheet, min_row=2, max_row=sheet.max_row, min_col=sheet.max_column, max_col=sheet.max_column)
    chart = BarChart()
    chart.add_data(selected_values)
    sheet.add_chart(chart, 'd13')

    wb.save(filename)
```
