# Day 3: Advance Data Types

List<br>

**List**: Ordered, Mutable (can be updated), duplications allowed
nums=[1,2,3,4,4]
nums[1]=87
nums.append(33)

**Range**: Ordered, Immutable, Arithmetic Sequence
a=range(2,50,2)
length=len(a)

**Tuple**: Ordered, Immutable, Allow duplication
nums=(1,2,3)

**Set**: Unordered, Mutable, Unique elements only
names={"Abdullah", "Afridi", "Khan"}
names.add("Affan")
names.remove("Khan")

**Dictionary**: A dictionary is a collection of key-value pairs. Each key is unique and maps to a value. Dictionaries are unordered, mutable, and indexed by keys.

```python
# List Data Type

list1 = [3,245,31,1,2342,23]
list2 = ["apple", "banana", "cherry"]

print("List 1: ", list1)
print("List 2: ", list2)

for items in list1:
    print(items)

list1.sort() # Sorting the list
print("Sorted List 1: ", list1)

list1.reverse() # Reversing the list
print("Reversed List 1: ", list1)

list2.append("orange") # Adding an item to the list
print("Updated List 2: ", list2)

list2.remove("banana") # Removing an item from the list
list2.pop(0) # Removing an item by index
list2.clear() # Removing all items from the list
print("Cleared List 2: ", list2)

list3 = list1 + list2 # Concatenating two lists
print("Concatenated List: ", list3)

list4 = list1 * 2 # Repeating a list
print("Repeated List: ", list4)

```

```python
# Tuple Data Type
tuple1 = (3,245,31,1,2342,23)
tuple2 = ("apple", "banana", "cherry")

print("Tuple 1: ", tuple1)
print("Tuple 2: ", tuple2)

for items in tuple1:
    print(items)

# We can not modify a tuple as it is immutable
```

```python
# Range Data Type
range1 = range(10) # Range from 0 to 9
range2 = range(1, 11) # Range from 1 to 10
range3 = range(0, 20, 2) # Range from 0 to 19 with a step of 2 (even numbers)
print("Range 1: ", list(range1)) # Converting range to list for display
print("Range 2: ", list(range2)) 
print("Range 3: ", list(range3)) 

for i in range(5):
    print("Number: ", i)

for a in range1:
    print("Range 1 Number: ", a)

# Different things we can do with range
print("Length of Range 1: ", len(range1)) # Length of the range
print("Minimum of Range 1: ", min(range1)) # Minimum value in the range
print("Maximum of Range 1: ", max(range1)) # Maximum value in the range
print("Sum of Range 1: ", sum(range1)) # Sum of all values in the range
print("Average of Range 1: ", sum(range1)/len(range1)) # Average of the range
```

```python
# Set Data Type
set1 = {3,245,31,1,2342,23}
set2 = {"apple", "banana", "cherry"}

print("Set 1: ", set1)
print("Set 2: ", set2)

for items in set1:
    print(items)

# Sets are unordered, so we cannot access elements by index
# print(set[2])  # This will raise an error

print("Length of Set 1: ", len(set1)) # Length of the set
print("Minimum of Set 1: ", min(set1)) # Minimum value in the set
print("Maximum of Set 1: ", max(set1)) # Maximum value in the set
print("Sum of Set 1: ", sum(set1)) # Sum of all values in the set
print("Average of Set 1: ", sum(set1)/len(set1)) # Average of the set
```

```python
# Dictionary Data Type
dict1 = {"name": "Abdullah", "age": 22, "city": "Peshawar"}
dict2 = {"fruit": "apple", "color": "red", "quantity": 5}

print("Dictionary 1: ", dict1)
print("Dictionary 2: ", dict2)

for key in dict1:
    print(f"{key}: {dict1[key]}")

print("Name: ", dict1["name"]) # Accessing value by key
print("Age: ", dict1["age"])

dict1["age"] = 23 # Modifying a value
print("Updated Age: ", dict1["age"])

dict1["country"] = "Pakistan" # Adding a new key-value pair
print("Updated Dictionary 1: ", dict1)

```

# Day 4: Operators & Type Casting

Arithmetic Operators<br>
Type Casting<br>

```python
# Arithmetic Operators
# Arithmetic operators are used to perform mathematical operations on numbers.

a = 10
b = 3
print("a + b = ", a + b) # Addition
print("a - b = ", a - b) # Subtraction
print("a * b = ", a * b) # Multiplication
print("a / b = ", a / b) # Division
print("a // b = ", a // b) # Floor Division: Quotient without the remainder
print("a % b = ", a % b) # Modulus: Remainder of the division
print("a ** b = ", a ** b) # Exponentiation
```

```python
# Type Casting
# Type casting is the process of converting one data type to another.

a="3"
b="4"
print("a + b = ", a + b) # This will concatenate the strings and give "34"
print(type(a))
print(type(b), "\n\n")

print("int(a) + int(b) = ", int(a) + int(b)) # This will convert the strings to integers and give 7
print("float(a) + float(b) = ", float(a) + float(b)) # This will convert the strings to floats and give 7.0
```

```python
number = "23"
number_int = int(number) # Converting string to integer
print(type(number))
print(type(number_int))
```

```python
a = 90
b = 23

print("Integer Sum: ", a + b) # This will give 113
print("Float Sum: ", float(a) + float(b)) # This will give 113.0
print("String sum: ", str(a) + str(b)) # This will give "9023" as it concatenates the strings
```

# Day 6: String Methods

String Methods in Python<br>
String Indexing

```python
# String methods
message = "Hello, world!"
print("Casefold: ", message.casefold()) #Covert the string to lowercase for case-insensitive comparisons
print("Upper: ", message.upper()) # Convert the string to uppercase
print("Lower: ", message.lower()) # Convert the string to lowercase
print("Capitalize: ", message.capitalize()) # Convert the first character to uppercase and the rest to lowercase
print("Title: ", message.title()) # Convert the first character of each word to uppercase and the rest to lowercase
print("Swapcase: ", message.swapcase()) # Lower to upper and upper to lower case
print("Center: ", message.center(85)) # Center the string within a field of a given width (85 in this case)
print("ljust: ", message.ljust(85)) # Left-justify the string within a field of a given width (85 in this case)
print("rjust: ", message.rjust(85)) # Right-justify the string within a field of a given width (85 in this case)

print("Count: ", message.count("o")) # Count the number of occurrences of a substring in the string
print("Startswith: ", message.startswith("Hello")) # Check if the string starts with a specific substring
print("Replace: ", message.replace(",", "*")) # Replace occurrences of a substring with another substring
print("Find 'w': ", message.find("w")) # Find the index of the first occurrence of a substring in the string, returns -1 if not found
print("Find 'world': ", message.find("world")) # Find the index of the first occurrence of a substring in the string, returns -1 if not found

print("Contains 'hello': ", "hello" in message) # Check if a substring is present in the string, returns True or False
print("Contains 'Hello': ", "Hello" in message)

print("Split: ", message.split(",")) # Split the string into a list of substrings based on a specified delimiter ("," in this case)

print(message[3]) # Accessing a character by index (0-based indexing)
print(message[0:5]) # Slicing the string from index 0 to 4 (5 is exclusive)
print(message[-6:]) # Slicing the last 6 characters of the string
```
