```python
print("Hello World") # String
print("This is a simple Python script.") # String
print(1234) # Integer
print(23.1) # Float
print("The sum of 5 and 6 is: ", 5+6) # String and Integer
print(True) # Boolean

```

```python
# Variables
first_name = "Abdullah"
last_name = "Afridi"
full_name = first_name + " " + last_name
print("Full Name: ", full_name) 

# OR

print("Full Name: ", first_name, last_name)


a=5
b=9
print("Sum of a and b is: ", a+b)

# Data Types
name = "Abdullah" # String
age = 25 # Integer
height = 5.11 # Float
is_student = True # Boolean

print(f"{name} is {age} years old, {height} feet tall, and is a student: {is_student}.") # This is formatted string (f-string) in Python.




```
